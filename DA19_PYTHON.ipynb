# Match case Statement

```python
# Match case is usef to write multiple if else conditions in easy understandable format

# Example-1

num = int(input("Enter a number:"))

match num:
    case 10:
        print("You entered 10")
    case 20:
        print("you entered 20")
    case 30:
        print("you entered 30")
    case 30:
        print("you entered 30000")
    case 50:
        print("you entered 50")
    case _:
        print("Wrong Number")
```

```python
# Example-2 : passing list to match

list1 = [10,20,30]

match list1:
    case [30,20,10]:
        print("Hi, How are you")
    case [10,20,30]:
        print("Go to Hell")
    case _:
        print("differnet list")
```

```python
# Example-3 : condition in match case

num  = 100

match num:
    case x if x < 50:
        print("Less then 50")
    case y if y > 100:
        print("Greater then 100")
    case z if z == 100:
        print("Equal to 100")
```

```python
n  = 100

match n:
    case n234 if n234 < 50:
        print("Less then 50")
    case _ if n > 100:
        print("Greater then 100")
    case _ if n == 100:
        print("Equal to 100")
```

```python
n1  = 100

match n1:
    case n12 if n12 < 50:
        print("Less then 50")
    case n12 if n12 > 100:
        print("Greater then 100")
    case n12 if n12 == 100:
        print("Equal to 100")
```

# Union operation

```python
set1 = {1,2,3,4,5}
set2 = {3,4,5,6,7}
set3 = {"Chinki", "Minki", 1,2,3,4}

set4 = set1 | set2
print(set4)

set5 = set1.union(set2)
print(set5)

# union on multiple sets
set6 = set1.union(set2, set3)
print(set6)

set7 = set1 | set2 | set3
print(set7)

# Inplace union operation : update , |=
# update() function will take the union of set1 and set2 and store result in set1
set1.update(set2)
print(set1)

set2 |= set3
print(set2)
```

```python
# Difference between union() and |

list1 = ['x', 'y', 'z']
tuple1 = (100, 200, 300)

set7 = set3.union(tuple1)
print(set7)
```

# Symmetric difference operation on set

```python
# Def: All elements of all the sets except common elements

set1 = {1,2,3,4,5}
set2 = {3,4,5,6,7}
set3 = {"Chinki", "Minki", 1,2,3,4}

set4 = set1 ^ set2
print(set4)

set5 = set1.symmetric_difference(set2)
print(set5)

set6 = set1 ^ set2 ^ set3
print(set6)

# symmetric_difference() takes exactly one argument
# set7 = set1.symmetric_difference(set2,set3) #  ERROR

set7 = set1.symmetric_difference(set2).symmetric_difference(set3)
print(set7)

# Inplace SD
set1.symmetric_difference_update(set2)
print(set1)

set2 ^= set3
print(set2)
```

```python
# Difference between symmetic_difference() and ^

list1 = ['x', 'y', 'z', 1,2,3]
tuple1 = (100, 200, 300, 1, 2, 3)
set10 = {100,1,2,3,4,5}

set11 = set10.symmetric_difference(list1)
print(set11)
```

# Decorator

```python
# Def: A decorator is a function that takes another function as input, and ADD extra functionalities
# without disturbing the original function functionality.
```

```python
# Example-1

def decorator_fun(cake):
    def wrapper_fun():   # wrapper function will add the extra functionality to our function
        print("0: Please collect the ingredients")
        cake()
        print("4 : Add strawbarry on the top")
    return wrapper_fun

def cake():   # this is our function which we want to decorate
    print("1. Mix the ingredients")
    print("2. Put it into Oven")
    print("3. Hurray : Cake is ready")

final_cake = decorator_fun(cake)   # passing our function to decorator function to add extra functionality
final_cake()
```

## @ symbol is used to define wrapper function

```python
# By using @ we can automatically applies the decorator 

# Example-2

def decorator_fun(cake):
    def wrapper_fun():   # wrapper function will add the extra functionality to our function
        print("0: Please collect the ingredients")
        cake()
        print("4 : Add strawbarry on the top")
    return wrapper_fun

@decorator_fun
def cake():   # this is our function which we want to decorate
    print("1. Mix the ingredients")
    print("2. Put it into Oven")
    print("3. Hurray : Cake is ready")

# final_cake = decorator_fun(cake)   # passing our function to decorator function to add extra functionality
# final_cake()

cake()
```

```python
# Example-3 : MRO (Method Resolution order)
class India:
    def __init__(self):
        print("This is India constructor function")

class Tel:
    def __init__(self):
        print("This is Telengana constructor function")
        super().__init__()  # calling the AP constructor function
    def display(self):
        print("Welcome to Telengana")

class AP:
    def __init__(self):
        print("This is AP constructor function")
        super().__init__()
    def show(self):
        print("Welcome to AP")

class HYD(Tel, AP, India):
    def __init__(self):
        print("This is HYD constructor function")
        super().__init__()
    def hello(self):
        super().display()
        super().show()

obj = HYD()

# printing MRO (Method Resolution Order)
print([i.__name__ for i in HYD.__mro__])
```

### overloading the arithmatic operators

```python
# Example-1 : overlaoded +
class Biryani:
    def __init__(self, variety, price):
        self.variety = variety
        self.price = price
    def __add__(self, second_obj):
        return self.price + second_obj.price    

dish1 = Biryani("Dum Biryani", 180)
dish2 = Biryani("Veg Biryani", 480)

print(dish1 + dish2)  # dish1.__add__(dish2)
```

```python
# Example-2 : overload -
class Biryani:
    def __init__(self, variety, price):
        self.variety = variety
        self.price = price
    def __sub__(self, second_obj):
        return self.price - second_obj.price   

dish1 = Biryani("Dum Biryani", 180)
dish2 = Biryani("Veg Biryani", 480)

print(dish1 - dish2)  # dish1.__add__(dish2)
```

```python
# Assignment Operator

# += : iadd
# -= : isub
# *= : imul
# /= : itruediv
# //= : ifloordiv
# %= : imod
# **= : ipow

class Biryani:
    def __init__(self, variety, price):
        self.variety = variety
        self.price = price
    def __iadd__(self, second_obj):
        self.price += second_obj.price  # dish1_price = dish1_price + dish2_price
        return self
    def __isub__(self, second_obj):
        self.price -= second_obj.price  # dish1_price = dish1_price + dish2_price
        return self
    def __imul__(self, second_obj):
        self.price *= second_obj.price  # dish1_price = dish1_price + dish2_price
        return self
    def __itruediv__(self, second_obj):
        self.price /= second_obj.price  # dish1_price = dish1_price + dish2_price
        return self
    def __ifloordiv__(self, second_obj):
        self.price //= second_obj.price  # dish1_price = dish1_price + dish2_price
        return self
    def __imod__(self, second_obj):
        self.price %= second_obj.price  # dish1_price = dish1_price + dish2_price
        return self
    def __ipow__(self, second_obj):
        self.price **= second_obj.price  # dish1_price = dish1_price + dish2_price
        return self
        
dish1 = Biryani("Dum Biryani", 180)
dish2 = Biryani("Veg Biryani", 480)

dish1 **= dish2   # dish1.__iadd__(dish2)  # dish1 = dish1 + dish2
print(dish1.price)
```

```python
# Unary operators
# - : neg
# ~ : invert

class Temp:
    def __init__(self, Ci, Ce):
        self.city = Ci
        self.celsius = Ce
    def __neg__(self):
        return -self.celsius
    def __invert__(self):
        return ~self.celsius

t1 = Temp("HYD", 41)
t2 = Temp("BLR", 38)

print(-t1) # t1.__neg__()
print(~t1) 
```

```python
# Overload the bitwise operator
# & : and
# | : or
# ^ : xor
# << : lshift
# >> : rshift

class Biryani:
    def __init__(self, variety, price):
        self.variety = variety
        self.price = price
    def __and__(self, second_obj):
        return self.price & second_obj.price  # dish1_price = dish1_price + dish2_price
    def __or__(self, second_obj):
        return self.price | second_obj.price  # dish1_price = dish1_price + dish2_price
    def __xor__(self, second_obj):
        return self.price ^ second_obj.price  # dish1_price = dish1_price + dish2_price
    def __lshift__(self, second_obj):
        return self.price << 2  # dish1_price = dish1_price + dish2_price
    def __rshift__(self, second_obj):
        return self.price >> 2  # dish1_price = dish1_price + dish2_price
        
dish1 = Biryani("Dum Biryani", 10)
dish2 = Biryani("Veg Biryani", 2)

print(dish1 >> dish2)   
```

# Data Abstraction

```python
# Def: Abstraction hides the implementation details and show only essential features.

# E.g. ATM machine

# Abstraction make the complexity of the project reduce.

# Rules of Abstraction in python:
#  1. Abstract class can have Abstract methods and Concrete methods
#  2. You cannot instantiate an abstract class
# 3. All subclasses must implement all the Abstract methods


# Abstract method: A method declared in Abstract class which must be overridden by the child class.
                # Abstract method need not to have the body part.
# Concrete method : This method can have body part. It is normal method.
```

```python
# Example-1 : Implement the Abstract class, abstract method, Concrete method
# 3. All subclasses must implement all the Abstract methods
#  2. You cannot instantiate an abstract class

from abc import ABC, abstractmethod

class country(ABC):
    def __init__(self):
        self.population = 150
    def display(self):           # concrete method
        print("India population is : " , self.population)
    @abstractmethod
    def flag(self):
        print("Hello, How are you")

    @abstractmethod
    def nationaltham(self):
        print("This is nationaltham")

class India(country):
    def flag(self):
        print("Tiranga")
        super().flag()
    def nationaltham(self):
        print("Jan Gan Man")
        

class Scotland(country):
     def flag(self):
        print("Square + Cross")
     def nationaltham(self):
        print("Flower of scotland")
        super().display()

obj1 = India()
obj2 = Scotland()

obj3 = country()

obj1.flag()
obj2.nationaltham()

obj1.display()
```

```python
# Example-1 : Implement the Abstract class, abstract method, Concrete method
# 3. All subclasses must implement all the Abstract methods
# 3.1 : Grandchild need not to have the abstract method

from abc import ABC, abstractmethod

class country(ABC):
    def __init__(self):
        self.population = 150
    def display(self):           # concrete method
        print("India population is : " , self.population)
    @abstractmethod
    def flag(self):
        print("Hello, How are you")

    @abstractmethod
    def nationaltham(self):
        print("This is nationaltham")

class India(country):
    def flag():
        pass
    def nationaltham(self):
        pass

class State(India):
    pass
        


obj1 = India()
obj2 = State()
```

```python
import re
```

```python
txt = """After every Exam 
Mom: Rakesh, See the son of Sharmaji
Rakesh: Sorry, Mom I just See the Daughtor of Sharma ji
"""
```

```python
# search: to search first matched substring in the text
obj = re.search("See", txt)
```

```python
# group(): to return only the founded substring
obj.group()
```

```python
# match(): check if the substring is present at the start of the text or not.
match_obj = re.match("After", txt)
match_obj
```

```python
# findall() : to find all the occurance of a substring in the txt
find_all = re.findall("See", txt)
find_all
```

```python
# sub(): to replace a substring with any other substring
sub_string = re.sub("Mom", "Dad", txt)
print(sub_string)
```

```python
# split() : break the sentence based upon the character
split_text = re.split(",", txt)

split_text
```
