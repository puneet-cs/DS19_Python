## Bitwise Operators ( & | ^ ~ << >>)

```python
# Why use the Bitwise operators:
    # 1. Bitwise operators are faster than arithmetic operators
    # 2. Many of the operations can be perform on Binary number easily as compare to decimal numbers
```

```python
# & | ^

a = 10
b = 3

print(a & b)
print(a | b)
print(a ^ b)
```

```python
# Bitwise Not ~

x = 1200

print(~x)
```

```python
y = 10

print(~y)
```

```python
a = 10
b = 3
c = 2
d = 8

print(a & b & c)
```

```python
print(a | b | c)
```

```python
print(a ^ b ^ c ^ d)
```

```python
# leftshift  << : add the bits (zero) to LSB side
# Left shift by 1 bit will double the value
# var * 2**bits

# Rightshift >> : Remove the bits from the LSB side
# Right shift by 1 bit will divide the value by 2
# floor(var / 2**bits)

var1 = 63

print(var1<<3)
```

```python
print(var1>>2)
```

# Assignment operators( +=, -=, *=, /=, %=, //=, &=, |=, <<=, >>=, :=)

```python
a = 10

a += 5
print(a)  # 15

a -= 5
print(a)  # 10

a *= 5
print(a)  # 50

a /= 5  # a = a/5
print(a)  # 10

a %= 5  # a = a%5
print(a)

a = 20

a //= 3
print(a)  # 6

a &= 3
print(a)  # 2

a |= 3
print(a)  # 3

a ^= 3
print(a)  # 

a = 10
a <<= 1
print(a)  # 20

a = 10
a >>= 1  # a = a>>1
print(a)  # 

print(x := 43)
```

# identity operators (is, is not)

```python
a = [1,2,3]
b = [1,2,3]
c = a

print(a == b)  # here we are comparing values, not objects
print(a is b)  # a and b are different objects
print(a is c)
print(a is not c)
```

# Output in Python

```python
# Different type of formatting for output

# 1. Concatenation using Commas
# 2. String concatenation using +
# 3. str.format()
# 4. f-string - Python 3.6+
# 5. Escape sequences in strings
# 6. End and Sep parameter
```

```python
# 1. Concatenation using Commas
name = "Puspa"
age = 41.2342342342
print("NAME:", name, "AGE:", age)
```

```python
# 2. String concatenation using +
print("NAME: "+ name + " " + "AGE: " + str(age))
```

```python
# 3. str.format()
print("Name: {} Age: {}".format(name, age))
```

```python
# 3. str.format()
print("Name: {1} Age: {0}".format(age, name))
```

```python
# 3. str.format()
print("Name: {v2} Age: {v1:.2f}".format(v1 = age, v2 = name))
```

```python
# 4. f-string - Python 3.6+
name = "Soubin"
age = 42

print(f"NAME: \"{name}\" \n AGE: {age}")
```

```python
# 6. End and Sep parameter

print(f"NAME:{name}", end = '@')
print(f"AGE:{age}", end = ' | ')
print("Hello Babai how is Pinni")
```

```python
print(name,age, sep = ' | ')
```
