# Match case Statement

```python
# Match case is usef to write multiple if else conditions in easy understandable format

# Example-1

num = int(input("Enter a number:"))

match num:
    case 10:
        print("You entered 10")
    case 20:
        print("you entered 20")
    case 30:
        print("you entered 30")
    case 30:
        print("you entered 30000")
    case 50:
        print("you entered 50")
    case _:
        print("Wrong Number")
```

```python
# Example-2 : passing list to match

list1 = [10,20,30]

match list1:
    case [30,20,10]:
        print("Hi, How are you")
    case [10,20,30]:
        print("Go to Hell")
    case _:
        print("differnet list")
```

```python
# Example-3 : condition in match case

num  = 100

match num:
    case x if x < 50:
        print("Less then 50")
    case y if y > 100:
        print("Greater then 100")
    case z if z == 100:
        print("Equal to 100")
```

```python
n  = 100

match n:
    case n234 if n234 < 50:
        print("Less then 50")
    case _ if n > 100:
        print("Greater then 100")
    case _ if n == 100:
        print("Equal to 100")
```

```python
n1  = 100

match n1:
    case n12 if n12 < 50:
        print("Less then 50")
    case n12 if n12 > 100:
        print("Greater then 100")
    case n12 if n12 == 100:
        print("Equal to 100")
```

# While loop

```python
i = 1
while i<=10:
    print(i)
    i += 1

print(i)
```

```python
i = 1
while i<=10:
    i += 1
    print(i)
```

```python
# break statement: to come out of the loop if a specific condition satisfy

i = 1
while(i<=10):
    i+=2
    print(i)
    if i == 5:
        break
```

```python
# Continue statement: Skip the next statements execution for one step
# continue statement will bring us back to the start of the loop
i = 1
while(i<=10):
    i+=2
    if i == 5:
        continue
    print(i)
```

```python
# else with while : else statement execute only if loop successfully complete
i = 1
while(i<=10):
    print(i)
    i+=1
else:
    print("Out of the loop")
```

```python
# pass statement : 

n = 1
while(n<=1000):
    if n == 100:
        pass
    n+=1
```

```python
# nested loop
i = 1
while(i<=3):
    j = 1
    while(j<=2):
        print(i,j)
        j+=1
    i+=1
```

```python
i = 1
while(i<=3):
    j = 1
    while(j<=2):
        if i == 1 or j == 2:
            break   
        print(i,j)
        j+=1
    i+=1
```

```python
# Practice question

i = 10
while i >= 1:
    i -= 1
    if i > 4 and i < 8:
        continue
    elif i < 2:
        break
    print(i)
```

```python
i = 1
j = 1
while i <= 10:
    if i % 3 != 0:
        j += 2
    elif j % 3 == 0:
        break
    i+=1

print(i+j)
```
