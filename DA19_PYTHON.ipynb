# Match case Statement

```python
# Match case is usef to write multiple if else conditions in easy understandable format

# Example-1

num = int(input("Enter a number:"))

match num:
    case 10:
        print("You entered 10")
    case 20:
        print("you entered 20")
    case 30:
        print("you entered 30")
    case 30:
        print("you entered 30000")
    case 50:
        print("you entered 50")
    case _:
        print("Wrong Number")
```

```python
# Example-2 : passing list to match

list1 = [10,20,30]

match list1:
    case [30,20,10]:
        print("Hi, How are you")
    case [10,20,30]:
        print("Go to Hell")
    case _:
        print("differnet list")
```

```python
# Example-3 : condition in match case

num  = 100

match num:
    case x if x < 50:
        print("Less then 50")
    case y if y > 100:
        print("Greater then 100")
    case z if z == 100:
        print("Equal to 100")
```

```python
n  = 100

match n:
    case n234 if n234 < 50:
        print("Less then 50")
    case _ if n > 100:
        print("Greater then 100")
    case _ if n == 100:
        print("Equal to 100")
```

```python
n1  = 100

match n1:
    case n12 if n12 < 50:
        print("Less then 50")
    case n12 if n12 > 100:
        print("Greater then 100")
    case n12 if n12 == 100:
        print("Equal to 100")
```
